```python
# Initial the map

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label
import folium

# 1. Define the launch site data based on your list
launch_sites = [
    {'name': 'CCAFS LC-40', 'lat': 28.562302, 'long': -80.577356},
    {'name': 'CCAFS SLC-40', 'lat': 28.563197, 'long': -80.576820},
    {'name': 'KSC LC-39A', 'lat': 28.573255, 'long': -80.646895},
    {'name': 'VAFB SLC-4E', 'lat': 34.632834, 'long': -120.610745}
]

# 2. Initialize the map (assuming nasa_coordinate is defined, e.g., [28.5623, -80.5773])
site_map = folium.Map(location=[28.5623, -80.5773], zoom_start=5)

# 3. For each launch site, add a Circle and a Marker
for site in launch_sites:
    # Add a circle to highlight the area
    circle = folium.Circle(
        [site['lat'], site['long']],
        radius=1000, # Radius in meters
        color='#d35400', 
        fill=True
    ).add_child(folium.Popup(site['name']))
    
    # Add a marker with the site name as a label
    marker = folium.map.Marker(
        [site['lat'], site['long']],
        icon=folium.DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site['name'],
        )
    )
    
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
from folium.plugins import MarkerCluster

# 1. Create a marker_color column in the dataframe
# Success (Class 1) = green, Failure (Class 0) = red
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# 2. Initialize the MarkerCluster
marker_cluster = MarkerCluster()

# 3. Add the cluster to the site_map
site_map.add_child(marker_cluster)

# 4. For each row in the dataframe, create a marker and add it to the cluster
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate
    # And customize the icon color based on the 'marker_color' column
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"Launch Site: {record['Launch Site']}\nStatus: {'Success' if record['class']==1 else 'Failure'}"
    )
    marker_cluster.add_child(marker)

# Display map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6371.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from folium.plugins import MousePosition

# Define the coordinate formatter for 5 decimal places
formatter = "function(num) {return L.Util.formatNum(num, 5);};"

# Initialize the MousePosition plugin
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

# Add the widget to your existing site_map
site_map.add_child(mouse_position)

# Display the map
site_map
# Save the map to an HTML file
site_map.save("spacex_launch_map2.html")

# In most environments, this will create a link you can click to open the map in a new tab
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
from folium.features import DivIcon

# Coordinates for the marker (The Coastline point)
coastline_coordinate = [coastline_lat, coastline_lon]

# Create the distance marker
distance_marker = folium.Marker(
    coastline_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Add it to the map
site_map.add_child(distance_marker)

# Add the line connecting the launch site to the coastline
lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], coastline_coordinate], weight=1)
site_map.add_child(lines)

# Show the map
site_map
# Save the map to an HTML file
site_map.save("spacex_launch_map3.html")

# In most environments, this will create a link you can click to open the map in a new tab
```
